### Get tweet_id to url mapping:

```python
tweets_urls = pd.read_csv("../../data/tweetid_url_map.csv", delimiter="\t", header=None, names= ["tweet_id", "url"])
```

### Get tweets with states:

```python
cursor.execute('Select tweet_id, state, country from newsTweets_location where state <> ""')
location_tweet_data = cursor.fetchall()
```

```python
len(location_tweet_data)
```

```python
location_tweet_data[:10]
```

### Merge

```python
tweet_url_df = pd.DataFrame(columns=['tweet_id', 'country', 'state', 'url'])
for row in location_tweet_data:
    urls = tweets_urls[tweets_urls.tweet_id == row[0]].url.unique()
    urls_len = len(urls)
    if not urls_len:
        continue
    tweet_url_df = tweet_url_df.append(
        pd.DataFrame(
            {"url":urls, "tweet_id":[row[0]]*urls_len, "country":row[2]*urls_len, "state" : row[1]*urls_len}))
```
