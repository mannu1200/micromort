### Clean the tweets

```python
def preprocess_text(str, level=2):
   r = requests.get("http://172.29.33.45:8000/textpreprocessor/{}?text={}".format(level, urllib.quote_plus(str)))
   return r.json()
```

```python
cleaned_tweets = []
for tweet in tweets:
    clean_json = preprocess_text(tweet)
    _str = ""
    for token in clean_json['token_list']:
       if token["core"]["class"] in [10]:
            _str += token["core"]["token"] + " "
    cleaned_tweets.append(_str.strip())
```

```python
for index in range(0,100):
    print "Original tweet:\n" + str(tweets[index]) + "\nCleaned :\n" + str(cleaned_tweets[index]) + "\n\n\n"
```

#### Save the cleaned tweets:

```python
f = open("../cleaned_tweets", "w")
for tweet in cleaned_tweets:
    f.write("%s\n" % tweet)
```

### load the pre trained model

```python
for tweet in tweets:
    labels = classifier.predict_single(cleaned_tweets[0], with_label=True)
    #if labels[0] != 'not_applicable':
    print tweet + " :::: " + str(labels) + "\n\n\n"
```
